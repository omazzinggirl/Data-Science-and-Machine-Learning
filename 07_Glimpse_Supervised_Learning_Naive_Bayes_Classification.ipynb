we will use an extremely simple generative model known as Gaussian naive Bayes, which proceeds by assuming each class is drawn from an axis-aligned Gaussian distribution. Because it is so fast and has no hyperparameters to choose, Gaussian naive Bayes is often a good model to use as a baseline classification,

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(data = iris, hue = 'species', height = 1.5)
```

```python
X = iris.iloc[:, :-1]
X.shape
```

```python
y = iris.iloc[:,-1] # 1D array is expected
y.shape
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train, X_test, y_train, y_test = tts(X, y, random_state = 1)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
